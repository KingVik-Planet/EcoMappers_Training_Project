# Labels

```python
x_dir = np.array([15, 4, 6,10 ,55,13])
y_dir = np.array([5, 9, 1, 0 ,5,3])

plt.plot(y_dir, linestyle = "dashed")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is my title")
plt.show()
```

```python
x_dir = np.array([15, 4, 6,10 ,55,13])
y_dir = np.array([5, 9, 1, 0 ,5,3])

font_1 = {"family": "serif", "color" : "blue", "size" : 20}
font_2 = {"family": "serif", "color" : "black", "size" : 10}

plt.title("This is my title", fontdict = font_1, loc = "center")
plt.xlabel("X axis", fontdict = font_2)
plt.ylabel("Y axis",fontdict = font_2)



plt.plot(y_dir, linestyle = "dashed")

plt.show()
```

# Grid

```python
x_dir = np.array([15, 4, 6,10 ,55,13])
y_dir = np.array([5, 9, 1, 0 ,5,3])

plt.plot(y_dir, linestyle = "dashed")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is my title")
plt.grid(linewidth = 1, linestyle = "--", color = "red")
plt.show()
```

# Scatter

```python
x = np.array([1,4,2,6,7,89])
y = np.array([3,6,6,6,7,9])
plt.scatter(x, y)
plt.show()
```

```python
x = np.array([1,4,2,6,7,89])
y = np.array([3,6,6,6,7,9])
plt.scatter(x, y, color  = "red")


x = np.array([10,14,12,16,70,189])
y = np.array([30,46,26,61,17,19])
colors = np.array(["green", "red", "blue", "yellow", "black", "orange"])

plt.scatter(x, y, c = colors)

plt.show()
```

```python
x = np.array([10,14,12,16,70,189])
y = np.array([30,46,26,61,17,19])
colors = np.array([0, 5,10,15,20,25])

plt.scatter(x, y, c = colors, cmap = "rainbow")
plt.colorbar()

plt.show()
```

```python
x = np.array([10,14,12,16,70,189])
y = np.array([30,46,26,61,17,19])
colors = np.array([0, 5,10,15,20,25])
sizes = np.array([2000,114,236,122,110,10])

plt.scatter(x, y, c = colors, cmap = "rainbow", s = sizes)
plt.colorbar()

plt.show()
```

```python
x = np.array([10,14,12,16,70,189])
y = np.array([30,46,26,61,17,19])
colors = np.array([0, 5,10,15,20,25])
sizes = np.array([2000,114,236,122,110,10])

plt.scatter(x, y, c = colors, cmap = "rainbow", s = sizes, alpha = 1)
plt.colorbar()

plt.show()
```

```python
import random

a = np.random.randint(200, size = (100))
b = np.random.randint(200, size = (100))
colors = np.random.randint(200, size = (100))
sizes = 20 * np.random.randint(200, size = (100))

plt.scatter(a, b, c = colors, s = sizes, alpha = 0.8, cmap = "rainbow")
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
x = ["Aimee","Lydia", "Leoncie", "Honore", "Farida"]
y = [30,46,26, 45, 56]
bar_label = ["r", "g", "b", "o", "c"]
bar_colors = ["tab:red", "tab:blue", "tab:green", "tab:orange", "tab:cyan"]

ax.bar(x, y, label = bar_label, color = bar_colors)

plt.show()
```

```python
y = np.array([30,46,26,61,17,19])
x = np.array(["Aimee","Lydia", "Leoncie", "Farida", "Balaka", "Honore"])
explo = [0.5,0,0,0,0,0]

plt.pie(y, labels = x, startangle = 90, explode = explo, shadow = True)
plt.show()
```

```python
y = np.array([30,46,26,61,17,19])
x = np.array(["Aimee","Lydia", "Leoncie", "Farida", "Balaka", "Honore"])
explo = [0.5,0,0,0,0,0]
colx = ["r", "b", "k", "g", "y","r"]

plt.pie(y, labels = x, startangle = 90, explode = explo, shadow = True,colors = colx)
plt.legend(title = "OurShare")
plt.show()
```
