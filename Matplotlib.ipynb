# Plotting Muliple line

```python
y_dir1 = np.array([5, 9, 1, 0 ,5,3])
y_dir2 = np.array([15, 4, 3, 10 ,6,9])

plt.plot(y_dir1, linestyle = "dashed", c = "y", linewidth = 1)
plt.plot(y_dir2, linestyle = "dashed")

plt.show()
```

```python
x_dir1 = np.array([15, 4, 6,10 ,55,13])
x_dir2 = np.array([15, 54, 23, 10 ,61,90])
y_dir1 = np.array([5, 9, 1, 0 ,5,3])
y_dir2 = np.array([15, 4, 3, 10 ,6,9])

plt.plot(x_dir1, x_dir2, y_dir1, y_dir2)


plt.show()
```

# Labels

```python
x_dir = np.array([15, 4, 6,10 ,55,13])
y_dir = np.array([5, 9, 1, 0 ,5,3])

plt.plot(y_dir, linestyle = "dashed")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("This is my title")
plt.show()
```

```python
x_dir = np.array([15, 4, 6,10 ,55,13])
y_dir = np.array([5, 9, 1, 0 ,5,3])

font_1 = {"family": "serif", "color" : "blue", "size" : 20}
font_2 = {"family": "serif", "color" : "black", "size" : 10}

plt.title("This is my title", fontdict = font_1, loc = "center")
plt.xlabel("X axis", fontdict = font_2)
plt.ylabel("Y axis",fontdict = font_2)



plt.plot(y_dir, linestyle = "dashed")

plt.show()
```
